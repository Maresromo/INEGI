```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f = '../data/raw/enigh2024_ns_viviendas_csv/viviendas.csv'
VIVIENDA = pd.read_csv(f, 
                       usecols = ['folioviv','aire_acond','entidad','factor']
                      )
VIVIENDA
```

```python
VIVIENDA.groupby("aire_acond")["factor"].sum()
```

```python
totales = VIVIENDA.groupby("aire_acond")["factor"].sum()

porcentajes = totales / totales.sum() * 100
print(porcentajes)
```

```python
# Ponderado
cuart_dorm_counts = VIVIENDA.groupby("aire_acond")["factor"].sum()

plt.figure(figsize=(8,5))
plt.bar(cuart_dorm_counts.index.astype(str), cuart_dorm_counts.values)
plt.xlabel("Nmero de dormitorios (cuart_dorm)")
plt.ylabel("Frecuencia ponderada")
plt.title("Histograma ponderado de dormitorios")
plt.show()
```
